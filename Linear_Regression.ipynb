```python
from google.colab import files
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
csv_data = uploaded_dict['headbrain .csv']
```

```python
df = pd.read_csv(io.BytesIO(csv_data))
print(uploaded_dict.keys())
print(df.head())
```

```python
x = df["Head Size(cm^3)"].values
y = df["Brain Weight(grams)"].values
x.shape
y.shape
```

```python
mean_x = np.mean(x)
print(mean_x)
```

```python
mean_y = np.mean(y)
m = len(x)
print(m)
```

```python
num = 0
denom = 0
for i in range(m):
  num += (x[i]-mean_x)*(y[i]-mean_y)
  denom += (x[i]-mean_x)**2
  b1 = num/denom
  b0 = mean_y - (b1*mean_x)
print(b1,b0)
```

```python
max_x = np.max(x) + 100
min_x = np.min(x) - 100
x = np.linspace(min_x,max_x,1000)
y = b0 + b1*x
print(y.shape)
```

```python
plt.plot(x,y, color = "red", label = "Regression Line")
plt.scatter(x,y,color = "blue", label = "Scatter Plot", s = 1)
plt.xlabel("Head Size in cm^3")
plt.ylabel("Brain Weight in grams")
plt.legend()
plt.grid(True)
plt.title("Linear regression for predicting brain weight based on head size")
plt.rcParams["figure.figsize"] = (20.0,10.0)
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
  y_pred = b0+b1*x[i]
  ss_t += (y[i] - mean_y)**2
  ss_r += (y[i] - y_pred)**2
  r2 = 1-(ss_r/ss_t)
print(r2)
```
