```python
from google.colab import files
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
uploaded_dict = files.upload()
csv_data = uploaded_dict['User_Data.csv']
```

```python
df = pd.read_csv(io.BytesIO(csv_data))
print(uploaded_dict.keys())
print(df.head())
```

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def gradient_descent(x, y, theta, alpha, iterations):
    m = len(y)
    for _ in range(iterations):
        z = np.dot(x, theta)
        h = sigmoid(z)
        gradient = np.dot(x.T, (h - y)) / m
        theta -= alpha * gradient
    return theta

def cost_function(x, y, theta):
    m = len(y)
    h = sigmoid(x @ theta)
    return -(1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))


def predict(x,theta):
  return(sigmoid(x @ theta)>= 0.5).astype(int)

def accuracy(y_true, y_pred):
  return np.mean(y_true == y_pred)

def predict(X, theta):
    theta = np.zeros((X.shape[1],1))
    z = np.dot(X, theta)
    return sigmoid(z)


if 'User ID' in df.columns:
    df.drop('User ID', axis=1, inplace=True)


gender_mapping = {"Male":1, "Female":0}
df["Gender"] = df['Gender'].map(gender_mapping)

df.iloc[:, :3] = (df.iloc[:, :3] - df.iloc[:, :3].mean()) / df.iloc[:, :3].std()


X = np.c_[np.ones((df.shape[0], 1)), df.iloc[:, :3].values]
y = df.iloc[:, 3].values.reshape(-1, 1)


def train_test_split(X, y, test_size=0.25, random_state=None):
    if random_state:
        np.random.seed(random_state)
    indices = np.random.permutation(len(X))
    test_size = int(len(X) * test_size)
    test_indices = indices[:test_size]
    train_indices = indices[test_size:]
    X_train, X_test = X[train_indices], X[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]
    return X_train, X_test, y_train, y_test

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)


theta = np.zeros((X.shape[1], 1))
alpha = 0.01
iterations = 1000
theta = gradient_descent(x_train, y_train, theta, alpha, iterations)


y_pred = predict(x_test, theta)

y_pred_rounded = np.where(y_pred >= 0.5, 1, 0)

def confusion_matrix(y_true, y_pred):
    tp = np.sum(np.logical_and(y_true == 1, y_pred == 1))
    tn = np.sum(np.logical_and(y_true == 0, y_pred == 0))
    fp = np.sum(np.logical_and(y_true == 0, y_pred == 1))
    fn = np.sum(np.logical_and(y_true == 1, y_pred == 0))
    return np.array([[tn, fp], [fn, tp]])

cm = confusion_matrix(y_test, y_pred_rounded)

print('Confusion matrix\n\n', cm)
print('\nTrue Positives(TP) = ', cm[0, 0])
print('\nTrue Negatives(TN) = ', cm[1, 1])
print('\nFalse Positives(FP) = ', cm[0, 1])
print('\nFalse Negatives(FN) = ', cm[1, 0])

accuracy = ((cm[0,0]+ cm[1,1])/(cm[0,0]+cm[0,1]+cm[1,0]+cm[1,1]))

print('Accuracy:',accuracy)
fit_new_point = np.array([[1, 55, 2000]])
fit_new_point[:, 1:3] = (fit_new_point[:, 1:3] - df.iloc[:, 1:3].mean().values.reshape(1, -1)) / df.iloc[:, 1:3].std().values.reshape(1, -1)
new_pred = predict(fit_new_point, theta)
print('\nPrediction for new data point: [1 = Yes, 0 = No] = ', np.where(new_pred >= 0.5, 1, 0))



```
